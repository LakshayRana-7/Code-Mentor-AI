```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import(
    accuracy_score,
    confusion_matrix,
    classification_report
)
```

```python
df= pd.read_csv(r"/content/Iris_Data - Iris_Data.csv")
```

```python
print(df.head())

x=df.drop("species", axis=1)
y=df["species"]
```

```python
print("\nFeatures")
print(x.head())

print("\nTarget")
print(y.head())
```

```python
x_train, x_test, y_train, y_test= train_test_split(
    x,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print("\nTraining samples:", x_train.shape[0])
print("\nTesting samples:", x_test.shape[0])
```

```python
model= LogisticRegression()
model.fit(x_train,y_train)
print("\nModel Trained Successfully")
print("\nIntercept")
print(model.intercept_)
print("\nCoefficients")
print(model.coef_)
y_pred=model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy:")
print(accuracy)

print("accuracy percentage=", accuracy*100)
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import pandas as pd
import matplotlib.pyplot as plt

# Confusion Matrix Plot
fig1, ax1 = plt.subplots(figsize=(6, 5)) # Create a figure and axes for the confusion matrix
ConfusionMatrixDisplay.from_predictions(y_true=y_test, y_pred=y_pred, cmap="Blues", ax=ax1)
ax1.set_title("Confusion Matrix") # Set title on the axes
plt.show() # Show the first figure

# Actual vs Predicted Plot
plt.figure(figsize=(8,5)) # Create a separate figure for the actual vs predicted plot

# Plot actual values
plt.plot(range(len(y_test)), y_test.factorize()[0],
         marker='o',
         label="actual")

# Plot predicted values
plt.plot(range(len(y_pred)), pd.Series(y_pred).factorize()[0],
         marker='x',
         label="predicted")

plt.xlabel("Sample Number")
plt.ylabel("Class")
plt.title("Actual vs Predicted")

plt.legend()
plt.show()
```

```python
new_flower = [[5.1, 3.5, 1.4, 0.2]]
predicted_species = model.predict(new_flower)
print("\nPrediction for New Flower:")
print(predicted_species)

score=model.score(x_test,y_test)
print("\nScore=",score)
```
